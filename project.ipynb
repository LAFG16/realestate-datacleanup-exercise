#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
prices = south_belt['price']

prices_mean = prices.mean()

prices_std = prices.std()

pricenorm = (prices - prices_mean) / prices_std

south_belt['normalized price'] = pricenorm

fig, axis = plt.subplots(2, 2, figsize=(10, 8))

for i, location in enumerate(south_belt['level5'].unique()):
    data = south_belt[south_belt['level5'] == location]
    axis[i//2, i%2].hist(data['normalized price'], bins = 30, alpha = 0.7)
    axis[i//2, i%2].set_title(f'Normalized price in {location}')
    axis[i//2, i%2].set_xlabel('Precio normalizado')
    axis[i//2, i%2].set_ylabel('inmuebles')
    

plt.tight_layout()
plt.show()
```

With these graphs, we can observe the following points:

Fuenlabrada: has a higher number of properties with an area between 50 and 100 square meters, and the price per square meter ranges between 1500 and 2000.
Alcorcón: has a higher number of properties with an area between 0 and 100 square meters, and the price per square meter ranges between 1500 and 2500.
Leganés: similar to Fuenlabrada, it has a higher number of properties with an area between 50 and 100 square meters, and the price per square meter ranges between 1500 and 2500.
Getafe: has a higher number of properties with an area between 0 and 200 square meters, and the price per square meter ranges between 1500 and 3000.

```python
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Stamen.Terrain)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:

south_belt['latitude'] = south_belt['latitude'].str.replace(",",".")
south_belt['longitude'] = south_belt['longitude'].str.replace(",",".")

icon1 = AwesomeIcon(
    name='home',
    marker_color='red',
    icon_color='darkred'
)
icon2 = AwesomeIcon(
    name='home',
    marker_color='green',
    icon_color='darkgreen'
)
icon3 = AwesomeIcon(
    name='home',
    marker_color='black',
    icon_color='gray'
)
icon4 = AwesomeIcon(
    name='home',
    marker_color='orange',
    icon_color='red'
)
icon_location = {
    'Leganés': icon1,
    'Fuenlabrada': icon2,
    'Getafe': icon3,
    'Alcorcón': icon4
}

for index, location in enumerate(south_belt['level5'].unique()):
    location_data = south_belt[south_belt['level5'] == location]
    
    for ind, row in location_data.iterrows():
        latitude = float(row['latitude'])
        longitude = float(row['longitude'])
        marker = Marker(location=(latitude, longitude), draggable=False, icon=icon_location[location])
        map.add_layer(marker)
```
